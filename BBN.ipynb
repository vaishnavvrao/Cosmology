## Introduction:
The theory of Big Bang Nucleosynthesis (BBN) effectively explains the origin of the $^4He$ abundance. When the temperature of the universe drops to around $10^{10}$ K—corresponding to the point where the weak interaction rate falls below the universe's expansion rate—the $n \leftrightarrow p$ processes freeze out. As a result, the neutron abundance remains nearly constant from this point onward, with only a slight decrease due to neutron decay. Shortly after neutron freeze-out, the temperature decreases further to approximately $10^9$ K, which breaks the "bottleneck" for deuterium synthesis. This initiates BBN, rapidly converting nearly all neutrons into $^4He$. Once neutrons are captured by $^4He$, they are no longer subject to neutron decay, preserving the helium abundance to the present day.

In this study, I first analyze the freeze-out abundance of neutrons by solving the interaction rate equation. Then, I evaluate how neutron decay affects neutron abundance over time. Following this, I determine the timing of BBN by considering the baryon-to-photon ratio and opening up of the "deuterium bottleneck". Finally, I calculate the $^4He$ abundance as a function of the baryon-to-photon ratio.


## The n:p ratio

In the early, radiation dominated Universe, neutrons and protons are interconverted by weak interactions:
$$n \longleftrightarrow p^+ + e^- + \bar{\nu_e}$$
$$n + \nu_e  \longleftrightarrow p^+ + e^- $$
$$n + e^+ \longleftrightarrow p^+ + \bar{\nu_e}$$

### Equilibrium Physics:

In the epoch where the interactions are fast compared to the expansion rate of the Universe, neutrons and protons maintain thermal equilibrium and their number densities are given by
$$n_{n,p} = 2\bigg(\frac{m_{n,p}T}{2\pi}\bigg)^{3/2} \exp\bigg[- \frac{m_{n,p}- \mu_{n,p}}{T}\bigg]$$
Defining $Q \equiv m_n - m_p= 1.294$ MeV, using $m_n/m_p \approx 1$ and $\mu_n - \mu_p = 0$, we can write the ratio of neutron to proton number densities as
$$\frac{n_n}{n_p} = \exp\bigg(-\frac{Q}{T}\bigg)$$
Thus, at high temperatures ($T \gg Q$) and in thermal equilibrium, there are the same number of neutrons as protons. As $T$ drops below $Q$, the neutron-to-proton ratio drops exponentially, and if they are in thermal equilibrium, all the neutrons go away.

### Non-Equilibrium Physics:

The real Universe is, however, not in thermal equilibrium. Following MBW Eq. 3.162, the rate of change of number densities of neutrons is:
$$\frac{dn_n}{dt} + 3H n_N = \lambda_{pn} n_p - \lambda_{np} n_n$$
where $\lambda_{np}$ is the rate of neutron to proton conversion and $\lambda_{pn}$ is the rate of proton to neutron conversion (units: s $^{-1}$). Let's assume all baryons are either neutrons or protons, and define $X_n$ as the fraction of baryons in neutrons. The fraction of protons is thus $X_p = 1 - X_n$. If we divide the above equation by the number density of baryons, we get
$$\frac{1}{n_b}\frac{dn_n}{dt} + 3 H X_n = \lambda_{pn} n_p - \lambda_{np} n_n$$
Noting that 
$$\frac{1}{n_b}\frac{dn_n}{dt} = \frac{dX_n}{dt} - 3HX_n$$
The rate of change of $X_n$ is given by
$$\dot{X_n} = -\lambda_{np}X_n + \lambda_{pn}(1-X_n)$$
where $\lambda_{np}$ is the rate of neutron to proton conversion and $\lambda_{pn}$ is the rate of proton to neutron conversion (units: s $^{-1}$). Since $\dot{X_n} = 0$ in thermal equilibrium, and $X_n/(1-X_n) = \exp(-Q/T)$, we can write
$$\lambda_{pn} = \lambda_{np}\exp\bigg(-\frac{Q}{T}\bigg)$$
and 
$$\dot{X_n} = \lambda_{np}[-X_n + e^{-Q/T}(1-X_n)]$$

Further, we define $x \equiv Q/T$ and re-write LHS as ($dX_n/dx$)($dx/dt$). Because $T \propto 1/a$, we have
$$\frac{1}{T}\frac{dT}{dt} = -H = -\sqrt{\frac{8\pi G \rho}{3}}$$

Thus, the rate equation becomes
$$\frac{dX_n}{dx}= \frac{\lambda_{np}}{x H }[-X_n + e^{-x}(1-X_n)]$$

### Weak interaction rate:
Reference: [Osterweil 2009](https://www.sas.rochester.edu/pas/assets/pdf/undergraduate/non-equilibrium_-universal_evolution.pdf)

If we ignore neutron decay, an approximation of the weak interaction rate of conversion of neutrons to protons in terms of $x$ is given by:
$$\lambda_{np} = \frac{255}{\tau_n x^5}(12+6x+x^2)$$
where $\tau_n = 887$ s is the neutron lifetime.

### Hubble rate:

As nucleosythesis occurs in the radiation era, $\rho$ largely comes from relativistic particles, so
$$\rho = \frac{\pi^2}{30}g_*T^4$$
where $g_*$ is the effective relativistic degrees of freedom, which is dependent on temperature. For the period of interest
$$g_* = g_\gamma + \frac{7}{8}(g_{e^{-}} + g_{e^{+}} + 3g_{\nu} + 3g_{\bar{\nu}}) = \frac{43}{4}= 10.75$$
Thus,
$$H(x) = \bigg(\frac{8\pi^3 G g_* Q^4}{90}\bigg)^{1/2} \frac{1}{x^2} \simeq \frac{1.13}{x^2} s^{-1}$$

### Solving the rate equation and neutron freeze out:
Numerically solving for $X_n(x)$, we get 
$$\lim_{x\to \inf} X_n = 0.148$$
Now,
$$X_{n,eq} = \frac{1}{1+e^x}$$
Therefore, $x_{f.o}= X_{n,eq}^{-1}(0.148) \simeq 1.7473$, which corresponds to $T_{f.o} \simeq 0.74$ MeV, which agrees well with the values given in MBW. The time at which this occurs can be calculated through
$$t_{f.o} = 1.3\,\text{s}\, \bigg(\frac{T}{1\text{MeV}}\bigg)^{-2} \bigg(\frac{g_*}{g_{*0}}\bigg)^{-2/3}$$
Using $g_{*0} = g_\gamma =2$, we get $t_{f.o} \simeq 0.77$ s.

```python
from headers import *
import astropy.constants as const
from scipy.integrate import odeint
```

```python
#Weak interaction rate:
def l_np(x):
    #Returns rate of interaction in $s^{-1}$ for a given x
    tau = 887 #s
    return 255*(12+ 6*x + x**2)/(tau*x**5)

#Define H as a function of x:
def H(x):
    #Returns Hubble rate in s^-1 for a given x
    return 1.13/x**2 

#Define the rate equation:
rate1 = lambda X, x: (l_np(x)/(x*H(x)))* (np.exp(-x)*(1-X) - X)
```

```python
#Make list of x-values:
x_list = 10**np.linspace(-6, 2, 100000+1)

#Numerically solve the rate equation:
Xn = np.squeeze(odeint(rate1, 0.5, x_list))
Xn_eq = 1/(1+np.exp(x_list))
```

```python
# make plot
fig, ax = plt.subplots()

ax.plot(np.log10(x_list), np.log10(Xn), lw=2, c='k', label=r'$X_{\rm n}$')
ax.plot(np.log10(x_list), np.log10(Xn_eq), lw=2, c='k', ls='--', label=r'$X_{\rm n,eq}$', alpha=0.5)

ax.set_xlabel(r'$\log_{10}x$')
ax.set_ylabel(r'$\log_{10}X$')

ax.set_xlim(-1,1.5)
ax.set_xticks([-1,-0.5,0,0.5,1,1.5])

ax.set_ylim(-1,-0.2)
ax.set_yticks([-1,-0.8,-0.6,-0.4,-0.2])

ax.legend()
plt.show()
```

## Effect of neutron decay on the n:p ratio:

So far we have not included neutron decay in the rate equations:
$$n \longrightarrow p^+ + e^- + \bar{\nu_e}$$
Thus, we add another decay term to the rate equation of neutrons
$$\dot{X_n} = -\lambda_{np}X_n + \lambda_{pn}(1-X_n) - X_n/\tau_n$$
where $\tau_n = 887$ s is the neutron lifetime. In terms of $x$, this becomes
$$\frac{dX_n}{dx}= \frac{\lambda_{np}}{x H }[-X_n + e^{-x}(1-X_n)] - \frac{X_n}{x H\tau_n}$$

```python
#Neutron decay half-life:
tau_n = 887 #s

rate2 = lambda X, x: rate1(X, x) - X/(x*H(x)*tau_n)
Xn2 = np.squeeze(odeint(rate2, 0.5, x_list))
```

```python
# make plot
fig, ax = plt.subplots()

ax.plot(np.log10(x_list), np.log10(Xn), lw=2, c='k', label=r'$X_{\rm n, no decay}$')
ax.plot(np.log10(x_list), np.log10(Xn2), lw=2, c='r', label=r'$X_{\rm n}$')
ax.plot(np.log10(x_list), np.log10(Xn_eq), lw=2, c='k', ls='--', label=r'$X_{\rm n,eq}$', alpha=0.5)

ax.set_xlabel(r'$\log_{10}x$')
ax.set_ylabel(r'$\log_{10}X$')

ax.set_xlim(-1,1.5)
ax.set_xticks([-1,-0.5,0,0.5,1,1.5])

ax.set_ylim(-1,-0.2)
ax.set_yticks([-1,-0.8,-0.6,-0.4,-0.2])

ax.legend()
plt.show()
```

We see that even after freeze out, the neutron-to-proton ratio continues to decrease due to neutron decay. This is because the neutrons that were frozen out continue to decay into protons. The effect of neutron decay on the neutron-to-proton ratio is significant, and it is essential to include it in the calculations to obtain accurate results for the primordial helium abundance.

## Deuterium "Bottleneck":

Complex nuclei are formed as a result of nuclear interactions of the baryons. For instance, $^4 He$ could, in principle, be directly formed in many-body collisions; 2p + 2n →4 He. However, the number densities at the time when this reaction can take place are too low and its rate is negligible compared to the rate of expansion. Hence, the light complex nuclei can be produced only in a sequence of two-body reactions. The first step of this is deuterium (D) production:
$$n + p \longleftrightarrow D + \gamma$$

Since deuterium has the lowest binding energy of all nuclei in the nuclear reaction network, its production serves as a bottleneck to get nucleosynthesis started. In other words appreciable number of heavier nuclei with higher binding energies like $^4He$ do not form until the number of deuterium atoms rises to an appreciable amount. Once, this number has been reached, the subsequent nuclear reactions proceed rapidly, until all the neutrons are consumed an locked into $^4He$. 

The production of deuterium has a rate per free neutron larger than the expansion rate, thus deuterium nuclei are always produced in equilibrium in the temperature range of our interest (that is until nucleosynthesis starts). Following the steps in MBW Section 3.4.2, we have MBW equation 3.210 which gives the deuterium fraction:
$$X_D \approx 16.4 X_n X_p \eta \bigg(\frac{m_N}{T}\bigg)^{-3/2} \exp\bigg(\frac{B_D}{T}\bigg)$$
where $\eta \equiv n_b/n_\gamma$ is the baryon-to-photon ratio, $m_N$ is the nucleon mass, and $B_D = 2.224$ MeV is the binding energy of deuterium. MBW also tells us the temperature at which appreciable deuterium production starts is given by (equation 3.201):
$$T_D \sim \frac{B_D}{|\ln \eta| + \frac{3}{2}\ln(m_N/T)}$$
The RHS of the equation also has a temperature dependence, however since it is a natural logarithm, it is only a weak dependence, which evaluates out to $\ln(m_N/T) \approx 6$ (mass of a nucleon is $\sim1$ GeV and the temperature around this time is $T \sim 0.1$ MeV). We thus get
$$T_D \sim \frac{B_D}{|\ln \eta| + 6}$$
Equivalently, in terms of time:
$$t_D \simeq 1.3 \times (T_D/1\text{MeV})^{-2} $$
We can write this because we are in the radiation dominated era, where $T \propto 1/a$ and $a \propto t^{1/2}$, so $T \propto t^{-1/2}$ and $g_* = g_\gamma =2$ as electrons and positrons have frozen out.

## Helium-4 abundance:

At the temperature calculated above is the temperature at which all the subsequent nuclear reactions also begin and cause the rapid locking up of neutrons into $^4He$. This is because the subsequent reaction chain proceeds very quickly, because at this temperature all nuclei heavier than deuterium can possess high equilibrium abundances. The reaction chain stops at $^4He$ because of the instability of heavier elements having mass number $A=5$ and $A=8$. We can thus write the mass fraction of $^4He$ approximately as:
$$Y \equiv X_{^4\rm He} \approx \frac{4(n_n/2)}{n_n + n_p} = \frac{2(n_n/n_p)_D}{1+(n_n/n_p)_D} = 2X_{n,D}$$
where $(n_n/n_p)_D$ is the neutron-to-proton ratio and $X_{n,D}$ is the neutron fraction at $T=T_D$.

We can use the neutron decay equation after freeze-out to obtain $X_n$ at $T=T_D$
$$X_{n,D} = X_{n,f.o} \exp\bigg( \frac{-(t_D - t_{f.o})}{\tau_n}\bigg) = 0.148 \exp\bigg( \frac{-(t_D - t_{f.o})}{\tau_n}\bigg)$$

If we use $\eta = 6.21 \times 10^{-10}$ [(WMAP 2008)](https://iopscience.iop.org/article/10.1088/0067-0049/180/2/306), we get the following values:

- $T_D \sim 0.08$ MeV
- $t_D \sim 194$ s
- $X_{n,D} \sim 0.119$
- $Y \sim 0.238$
- $(n_n/n_p)_D \sim 0.135$

These values are very close to $Y = 0.2489 \pm 0.0002$ obtained by [Cybert 2008](https://iopscience.iop.org/article/10.1088/1475-7516/2008/11/012)

## Effect of varying baryon-photon ratio:


```python
#Define a function to calculate the temperature of nucleosynthesis onset (T_D):
def T_D(eta):
    #Returns the temperature of nucleosynthesis onset in MeV

    #Binding energy of deuterium in MeV:
    B_D = 2.224 #MeV

    return B_D/(np.abs(np.log(eta)) + 6)

#Define function to calculate time of nucleosynthesis onset (t_D):
def t_D(eta):
    #Returns the time of nucleosynthesis onset in s

    #Temperature of nucleosynthesis onset in MeV:
    T = T_D(eta)

    #Time of nucleosynthesis onset in s:
    t_D = 1.3/T**2

    return t_D

#Define a function to calculate the neutron-to-proton ratio at nucleosynthesis onset:
def X_n_D(eta):
    #Returns the neutron-to-proton ratio at nucleosynthesis onset
    tau_n = 887 #s
    t2 = t_D(eta)

    #Time of freeze out:
    t1 = 1.3/0.74**2 * (10.75/2)**(-2/3)

    X = 0.148*np.exp(-(t2-t1)/tau_n)
    return X

#Define a list of eta-values:
eta_list = 10**np.linspace(-11, -8, 100+1)

#Calculate the helium mass fraction for each eta-value:
X_n_D_list = np.array([X_n_D(eta) for eta in eta_list])
Y = 2 * X_n_D_list
```

```python
#Plot the helium mass fraction as a function of eta:
fig, ax = plt.subplots()

ax.plot(eta_list, Y, lw=2, c='k', label=r'$Y$')
ax.set_xlabel(r'$\eta$')
ax.set_ylabel('Primordial He Mass Fraction')
ax.set_xscale('log')
ax.set_xlim(10**(-11), 10**(-8))
ax.set_ylim(0.2,0.28)

ax.fill_between([6.05e-10,6.37e-10], [0,0], [1,1], 
                facecolor='r', edgecolor=None, alpha=0.2, label=r'$\rm WMAP\ 2008$')

ax.fill_between([10**(-12), 10**(-7)], [0.232,0.232], [0.258,0.258], 
                facecolor='k', edgecolor=None, alpha=0.2, label=r'$\rm Coc\&Vangioni\ 2010$')
ax.legend(
    loc='upper left'
)
plt.show()
```

```python
#Calculations for eta = 6.21e-10 (WMAP 2008) parameters at nucleosynthesis onset:
eta = 6.21e-10
temp = T_D(eta)
time = t_D(eta)
X = X_n_D(eta)
Y = 2*X

ratio = X/(1-X)

print(f'eta = {eta}')
print(f'T = {temp} MeV')
print(f't = {time} s')
print(f'X = {X}')
print(f'Y = {Y}')
print(f'N/P = {ratio}')
```

## Importance of measuring Y:
By measuring the abundance of priordial $^4He$ in the Universe (most of the $^4He$ in the Universe is primordial anyway) and other BBN elements through $D/H$ and $^3 He/H$, we can constrain $\eta$. $n_\gamma$ can be measured now from the CMB and hence $n_\gamma$ at the time of BBN can be obtained through the scale factot $a$. Thus, this is an independent way to constrain $n_b$ or $\Omega_b$ in the Universe.
